# Python e API

usando a API do site https://docs.awesomeapi.com.br/api-de-moedas, vou extrai um ano de cotação do BitCoin e exibir em um grafico, lembrando que esse site permite cotação de mais de 30 moedas diferentes

#### Pegar as cotações do BitCoin de Jan/20 a Out/20

```python
import requests
import json
```

```python
cotacaoBTC = requests.get('https://economia.awesomeapi.com.br/BTC-BRL/100?start_date=20200101&end_date=20201030')
cotacao_dicBTC = cotacaoBTC.json()
#print(cotacao_dicBTC)

lista_cotacaoBTC = [float(item['bid']) for item in cotacao_dicBTC]
lista_cotacaoBTC.reverse()
print(lista_cotacaoBTC)
```

#### Gráfico com as cotações do BitCoin

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(lista_cotacaoBTC)
plt.show()
```
